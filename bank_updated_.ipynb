```python
# Bagging for Decision tree

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

dt= DecisionTreeClassifier()

bag_clf= BaggingClassifier(estimator=dt,
                           n_estimators= 100,
                           max_samples= 0.6,
                           max_features= 0.7)

training_score=[]
testing_score=[]
testing_precision=[]
testing_recall=[]
testing_f1=[]


for i in range(1,100):
  x_train,x_test,y_train,y_test= train_test_split(x_resample, y_resample, test_size= 0.3, random_state= i)
  bag_clf.fit(x_train, y_train)
  y_pred_train= bag_clf.predict(x_train)
  y_pred_test= bag_clf.predict(x_test)
  training_score.append(accuracy_score(y_train, y_pred_train))
  testing_score.append(accuracy_score(y_test, y_pred_test))
  testing_precision.append(precision_score(y_test, y_pred_test))
  testing_recall.append(recall_score(y_test, y_pred_test))
  testing_f1.append(f1_score(y_test, y_pred_test))
print("Accuracy score of train:", np.mean(training_score).round(2))
print("Accuracy score of test:", np.mean(testing_score).round(2))
print("Precision score of test:", np.mean(testing_precision).round(2))
print("Recall score of test:", np.mean(testing_recall).round(2))
print("F1 score of test:", np.mean(testing_f1).round(2))
```

```python
import numpy as np
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Assuming x_resample and y_resample are your preprocessed features and target

# Fit and transform the data for preprocessing
scaler = StandardScaler()
pca = PCA(n_components=10)  # Use the same number of components as during training
imputer = SimpleImputer(strategy='mean')
ros = RandomOverSampler()

# Fit on training data
ss_x = scaler.fit_transform(x_resample)
x_pca = pca.fit_transform(ss_x)
x_imputed = imputer.fit_transform(x_pca)
x_resample_final, y_resample_final = ros.fit_resample(x_imputed, y_resample)

# Splitting data for evaluation
x_train, x_test, y_train, y_test = train_test_split(x_resample_final, y_resample_final, test_size=0.3, random_state=42)

# Train your BaggingClassifier
dt = DecisionTreeClassifier()
bag_clf = BaggingClassifier(estimator=dt,
                            n_estimators=100,
                            max_samples=0.6,
                            max_features=0.7)

training_score = []
testing_score = []
testing_precision = []
testing_recall = []
testing_f1 = []

for i in range(1, 100):
    x_train, x_test, y_train, y_test = train_test_split(x_resample_final, y_resample_final, test_size=0.3, random_state=i)
    bag_clf.fit(x_train, y_train)
    y_pred_train = bag_clf.predict(x_train)
    y_pred_test = bag_clf.predict(x_test)
    training_score.append(accuracy_score(y_train, y_pred_train))
    testing_score.append(accuracy_score(y_test, y_pred_test))
    testing_precision.append(precision_score(y_test, y_pred_test))
    testing_recall.append(recall_score(y_test, y_pred_test))
    testing_f1.append(f1_score(y_test, y_pred_test))

print("Accuracy score of train:", np.mean(training_score).round(2))
print("Accuracy score of test:", np.mean(testing_score).round(2))
print("Precision score of test:", np.mean(testing_precision).round(2))
print("Recall score of test:", np.mean(testing_recall).round(2))
print("F1 score of test:", np.mean(testing_f1).round(2))

# Define a function to predict bankruptcy based on input features
def predict_bankruptcy():
    # Take inputs from the user
    debt_ratio = float(input("Enter Debt Ratio: "))  # Collect input for debt ratio
    net_income_to_assets = float(input("Enter Net Income to Total Assets: "))  # Collect input for net income to assets
    net_worth_to_assets = float(input("Enter Net Worth to Total Assets: "))  # Collect input for net worth to assets

    # Transform input data using the complete preprocessing pipeline
    input_data = np.array([[debt_ratio, net_income_to_assets, net_worth_to_assets, 0, 0, 0, 0, 0, 0, 0, 0]])  # Pad with zeros to match 11 features
    input_data_scaled = scaler.transform(input_data)  # Transform using trained scaler
    input_data_pca = pca.transform(input_data_scaled)  # Transform using trained PCA
    input_data_imputed = imputer.transform(input_data_pca)  # Transform using trained Imputer

    # Predicting bankruptcy based on input features
    prediction = bag_clf.predict(input_data_imputed)
    return prediction

# Example usage:
prediction = predict_bankruptcy()
print("Prediction:", "Bankrupt" if prediction == 1 else "Not Bankrupt")
```
